```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('train.csv')
print(df.head())
```

```python
survived_counts=df['Survived'].value_counts()
plt.bar(survived_counts.index, survived_counts.values, color=['red','green'])
plt.xlabel('Survived')
plt.ylabel('Count')     
plt.title('Survival Counts on Titanic')
plt.show()
```

```python
sex_count=df['Sex'].value_counts()
plt.pie(sex_count, labels=sex_count.index, autopct='%1.1f%%', colors=['blue', 'pink'])
plt.title("Gender Distribution")
plt.show()
```

```python
plt.hist(df['Age'].dropna(),bins=30, color='purple', edgecolor='black')
plt.title('Age Distribution of Passengers')
plt.xlabel('Age')
plt.ylabel('Number of Passengers')
plt.show()
         
```

```python
import seaborn as sns
gender_survival=df.groupby(['Sex', 'Survived']).size().unstack()
gender_survival.plot(kind='bar', stacked=True, color=['red', 'green'])
plt.title('Survival by gender')
plt.xlabel('Number of passengers')
plt.xticks(rotation=0)
plt.legend(['Did not survive', 'Survived'])
plt.show()
```

```python
sns.boxplot(x='Pclass', y='Age', data=df)
plt.title('Age Distribution by Passenger class')
plt.xlabel('Passenger Class')
plt.ylabel('Age')
plt.show()
```
